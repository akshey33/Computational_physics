```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_style("whitegrid")

# Initialize lists to store data from all files
temps_list = []
energy_list = []

# Define colors and markers for each lattice size
colors = ['blue', 'orange', 'green', 'red', 'purple']
markers = ['o', 's', '^', 'D', 'x']

# Load data from files for different lattice sizes
for L in range(10, 26, 5):
    filename = f"energy_vs_temperature_at_L_{L}.txt"
    data = np.loadtxt(filename)
    temps_list.append(data[:, 0])
    energy_list.append(data[:, 1])

# Plot energy vs temperature for all lattice sizes
plt.figure(figsize=(10, 6))

for i in range(len(temps_list)):
    plt.plot(temps_list[i], energy_list[i], label=f'L = {10 + i * 5}', linestyle='--', color=colors[i]) # dashed lines
    plt.scatter(temps_list[i], energy_list[i], color=colors[i], marker=markers[i], label=f'L = {10 + i * 5}')

plt.xlabel('Temperature', fontsize=14)
plt.ylabel('Energy', fontsize=14)
plt.title('Energy vs. Temperature', fontsize=16)
plt.legend()

# Customize tick label font size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(False)
plt.show()
```
